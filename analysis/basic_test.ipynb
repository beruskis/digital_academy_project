```python
import pandas as pd

df = pd.read_csv(r'C:\SAPDevelop\digital_academy_project\MAIN_1990-2025forest-area-as-share-of-land-area.csv')

filtered = df[df['Year'].isin([1990, 2025])][['Entity', 'Year', 'Share of land covered by forest']]

table = filtered.pivot(index='Entity', columns='Year', values='Share of land covered by forest')
table.columns = ['Forest share 1990 (%)', 'Forest share 2025 (%)']
table = table.round(2)

table.to_csv(r'C:\SAPDevelop\digital_academy_project\test1.csv')
print('Saved to test1.csv')

table
```

```python
table['Change (%)'] = (table['Forest share 2025 (%)'] - table['Forest share 1990 (%)']).round(2)

top10 = table['Change (%)'].abs().nlargest(10)
table.loc[top10.index].sort_values('Change (%)', key=abs, ascending=False)
```
